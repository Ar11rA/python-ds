```python
# input
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1.0,1.8,3.3,4.5,6.3])
```

```python
# how data looks
import matplotlib.pyplot as plt

plt.plot(x, y, 'ro')
```

```python
# linear regression algo

import numpy as np

X = np.mean(x)
Y = np.mean(y)
N = len(y)

coeff = (np.sum(x * y) - N* X * Y)/(np.sum(x * x) - N * X * X)
intercept = Y - coeff * X
x_line = np.linspace(0,5,100)
y_line = coeff * x_line + intercept
plt.plot(x_line, y_line, '-r')
plt.plot(x, y, 'ro')
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression() 
x_reshaped = np.reshape(x, (-1,1))
print(x_reshaped)
regressor.fit(x_reshaped, y)

print(regressor.intercept_)
print(regressor.coef_)
y_pred = regressor.predict([[6]])
print(y_pred)
```

```python
# Example for sklearn LR

import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array([[1], [1], [2], [3]])
print(X)
# y = 2 * x_0 + 3
y = np.dot(X, np.array([2])) + 3
print(y)
reg = LinearRegression().fit(X, y)
reg.score(X, y)

reg.coef_

reg.intercept_

reg.predict(np.array([[4]]))
```

```python
# class example

class CustomLinearRegression:
  def __init__(self):
    self.coeff = 0.0
    self.intercept = 0.0

  def fit(self, x, y):
    X = np.mean(x)
    Y = np.mean(y)
    N = len(y)

    coeff = (np.sum(x * y) - N* X * Y)/(np.sum(x * x) - N * X * X)
    intercept = Y - coeff * X
    self.coeff = coeff
    self.intercept = intercept

  def predict(self, x):
    return self.coeff * x + self.intercept

lr = CustomLinearRegression()
lr.fit(x, y)
print(lr.coeff)
print(lr.intercept)
print(lr.predict(6))
```
